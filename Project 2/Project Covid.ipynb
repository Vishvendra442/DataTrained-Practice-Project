# Model Building

# Encoding

```python
#Encoding columns with catagorical data
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
ds['Province_State']=le.fit_transform(ds['Province_State'])
```

```python
#Converting Whole Dataset to same Data type
ds=ds.astype('int64')
```

# Standard Scaling

```python
#Normalising
from sklearn.preprocessing import StandardScaler

#Making X and Y variables
x=ds.drop('Deaths', axis=1)
y=ds['Deaths']

#Standard Scaling
#sc=StandardScaler()
#x=sc.fit_transform(x)
```

```python
#ML Models
from sklearn.linear_model import LinearRegression, Lasso, Ridge, ElasticNet
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor

#model selection
from sklearn.model_selection import train_test_split, GridSearchCV

#metrics
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
#testing Different Models
model=[LinearRegression(),SVR(), DecisionTreeRegressor(), KNeighborsRegressor(),Lasso(), Ridge(), ElasticNet()]

x_train,x_test,y_train,y_test=train_test_split(x, y, random_state=42, test_size=0.20)

for i in model:
    i.fit(x_train,y_train)
    pred=i.predict(x_test)
    print(i)
    print(i.score(x_train,y_train))
    print('R2 Score')
    print(r2_score(y_test,pred))
    print('MAE')
    print(mean_absolute_error(y_test,pred))
    print('MSE')
    print(mean_squared_error(y_test,pred))
    print()
    print('..................................................................................................................')
```

We are getting best R2 Score with Linear Regression

```python
#Calling Model
le=LinearRegression()

#Train Test Split
x_train,x_test,y_train,y_test=train_test_split(x, y, random_state=42, test_size=0.20)

#fitting Train data in model
le.fit(x_train,y_train)

#Predecting Values
pred=le.predict(x_test)

#printing Matrics
print(le.score(x_train,y_train))
print('R2 Score')
print(r2_score(y_test,pred))
print('MAE')
print(mean_absolute_error(y_test,pred))
print('MSE')
print(mean_squared_error(y_test,pred))
print()
```

```python
#Cross Valadation
from sklearn.model_selection import cross_val_score
cross_val_score(le,x,y,cv=5).mean()
```

```python
# Saving best Model
import joblib
joblib.dump(le,"covid.pkl")
```
