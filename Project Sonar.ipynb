```python
import pandas as pd
import numpy as np
ds=pd.read_csv('sonar.csv', header=None)
ds
```

```python
ds.describe()
```

```python
import matplotlib.pyplot as py
import seaborn as sns
```

```python
sns.heatmap(ds.isnull())
```

```python
from scipy.stats import zscore
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
ds[60]=le.fit_transform(ds[60])
z=abs(zscore(ds))
```

```python
print(np.where(z>3))
```

```python
ds=ds[(z<3).all(axis=1)]
```

```python
ds
```

```python
sns.countplot(ds[60])
```

as we see the data is evenly divided so accuracy metrics will give better result

```python
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
import warnings
warnings.filterwarnings('ignore')
```

```python
y=ds.iloc[:,-1:]
y
```

```python
x=ds.iloc[:,:-1]
```

```python
x
```

```python
pca=PCA(n_components=10)
x=pca.fit_transform(x)
```

```python
x.shape
```

```python
x=pd.DataFrame(data=x)
```

```python
x
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=sc.fit_transform(x)
```

```python
lr=LogisticRegression()
```

```python
max_ac_score=0

for r_state in range(42,100):
    x_train,x_test,y_train,y_test=train_test_split(x,y, random_state=r_state, test_size=0.20)
    lr.fit(x_train,y_train)
    predlr=lr.predict(x_test)
    acc_score=accuracy_score(y_test,predlr)
    if acc_score>max_ac_score:
        max_ac_score=acc_score
        final_rs=r_state
print('max accuracy score corrosponding to' ,final_rs ,'is', max_ac_score)
```

we will use random state = 42

```python
from sklearn.model_selection import GridSearchCV
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y, random_state=42, test_size=0.20)
```

```python
dtc=DecisionTreeClassifier()
```

```python

dtc_para= {'criterion': ['gini' , 'entropy']}

dtcgd=GridSearchCV( estimator=dtc ,param_grid=dtc_para ,scoring='accuracy' ,cv=5)
dtcgd.fit(x,y)
best_para=dtcgd.best_params_
best_score=dtcgd.best_score_
```

```python
print(best_para)
print(best_score)
```

```python
dtc=DecisionTreeClassifier(criterion='gini')
dtc.fit(x_train,y_train)
preddtc=dtc.predict(x_test)
print(dtc.score(x_train,y_train))
print(accuracy_score(y_test,preddtc))
print(confusion_matrix(y_test,preddtc))
print(classification_report(y_test,preddtc))
```

```python
sv=SVC()
sv_para={'kernel' : ['linear', 'rbf', 'poly'], 'C':[1,10]}
svgd=GridSearchCV(sv ,sv_para)
svgd.fit(x,y)
print(svgd.best_params_)
```

```python
svc=SVC(kernel='linear', C=10)
svc.fit(x_train,y_train)
predsvc=svc.predict(x_test)
print(svc.score(x_train,y_train))
print(accuracy_score(y_test,predsvc))
print(confusion_matrix(y_test,predsvc))
print(classification_report(y_test,predsvc))
```

```python
gn=GaussianNB()
gn.fit(x_train,y_train)
predgn=gn.predict(x_test)
print(gn.score(x_train,y_train))
print(accuracy_score(y_test,predgn))
print(confusion_matrix(y_test,predgn))
print(classification_report(y_test,predgn))
```

```python
rd=RandomForestClassifier(n_estimators=100, random_state=42)
rd.fit(x_train,y_train)
predrd=rd.predict(x_test)
print(rd.score(x_train,y_train))
print(accuracy_score(y_test,predrd))
print(confusion_matrix(y_test,predrd))
print(classification_report(y_test,predrd))
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knc=KNeighborsClassifier()
knc.fit(x_train,y_train)
print(knc.score(x_train,y_train))
predknc=knc.predict(x_test)
print(accuracy_score(y_test,predknc))
print(confusion_matrix(y_test,predknc))
print(classification_report(y_test,predknc))
```

```python
lr.fit(x_train,y_train)
print(lr.score(x_train,y_train))
predlr=lr.predict(x_test)
print(accuracy_score(y_test,predlr))
print(confusion_matrix(y_test,predlr))
print(classification_report(y_test,predlr))
```

```python
from sklearn.model_selection import cross_val_score

print(cross_val_score(svc,x,y,cv=5).mean())
print(cross_val_score(lr,x,y,cv=5).mean())
```

we see that svc model is working well so it will be our final model
