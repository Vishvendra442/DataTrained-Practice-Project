```python
business=ds[['Country', 'Economy']]
business=business.sort_values(by='Economy', ascending=False).head()
business.set_index('Country',inplace=True)
```

```python
gov=ds[['Country', 'Trust']]
gov=gov.sort_values(by='Trust', ascending=False).head()
gov.set_index('Country',inplace=True)
gov.plot(kind='bar')
```

```python
demo=ds[['Country', 'Freedom']]
demo=demo.sort_values(by='Freedom', ascending=False).head(5)
demo.set_index('Country',inplace=True)
demo.plot(kind='bar')
```
